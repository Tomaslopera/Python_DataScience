```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Cargar Dataset

```python
diamonds = sns.load_dataset('diamonds')
```

### Análisis Exploratorio

```python
diamonds.info()
```

### Visualizaciones

```python
# Relaciones entre variables
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamonds, x='carat', y='price', hue='cut', alpha=0.7)
plt.title('Relación entre Carat y Price agrupada por Cut')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid()
plt.show()
```

```python
# Distribuciones de variables numéricas
plt.figure(figsize=(10, 6))
sns.histplot(diamonds['price'], kde=True, bins=50, color='blue')
plt.title('Distribución de los precios de los diamantes')
plt.xlabel('Price')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()
```

```python
# Comparaciones de variables numéricas agrupadas por categoría
plt.figure(figsize=(15, 6))
sns.boxplot(data=diamonds, x='cut', y='price', palette='Set1', hue='color') # Hue se utiliza para agrupar por otra variable
plt.title('Distribución de precios por tipo de corte')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.legend(title='Color', bbox_to_anchor=(1.05, 1))
plt.grid()
plt.show()
```

### joinplot()

```python
sns.jointplot(data=diamonds, x='carat', y='price', hue='cut', height=8)
plt.suptitle('Relación entre Carat y Price con regresión', y=1.02)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.grid()
plt.show()
```

### pairplot()

```python
sns.pairplot(diamonds, hue='cut', diag_kind='kde', palette='husl', height=2.5)
plt.suptitle('Relaciones entre variables numéricas', y=1.02)
plt.show()
```
